```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load dataset
df = pd.read_csv("Chocolate_Sales 2 (1).csv")

# -------------------------
# Task 1: Data Quality Issues
# -------------------------
print("First 5 rows:")
print(df.head())

print("\nMissing values:")
print(df.isnull().sum())

print("\nDuplicates:", df.duplicated().sum())

print("\nData types before cleaning:")
print(df.dtypes)

# Fix data types
df["Amount"] = df["Amount"].replace({'\\$': '', ',': ''}, regex=True).astype(float)
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors="coerce")

print("\nData types after cleaning:")
print(df.dtypes)

# -------------------------
# Task 2: Missing Value Strategy
# -------------------------
print("\nTask 2:")
print("No missing values were found in the dataset.")
print("If missing values existed in numerical columns, mean imputation would be a suitable strategy.")

# -------------------------
# Task 3: Detect and Handle Outliers using IQR
# -------------------------
numerical_cols = ["Amount", "Boxes Shipped"]

Q1 = df[numerical_cols].quantile(0.25)
Q3 = df[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df[numerical_cols] < lower_bound) | (df[numerical_cols] > upper_bound))

print("\nOutliers in each numerical column:")
print(outliers.sum())

df_clean = df[~outliers.any(axis=1)]

print("\nShape before removing outliers:", df.shape)
print("Shape after removing outliers:", df_clean.shape)

# -------------------------
# Task 4: Normalization
# -------------------------
minmax_scaler = MinMaxScaler()
df_minmax = df_clean.copy()
df_minmax[numerical_cols] = minmax_scaler.fit_transform(df_clean[numerical_cols])

print("\nMin-Max Normalization:")
print(df_minmax[numerical_cols].head())

zscore_scaler = StandardScaler()
df_zscore = df_clean.copy()
df_zscore[numerical_cols] = zscore_scaler.fit_transform(df_clean[numerical_cols])

print("\nZ-score Normalization:")
print(df_zscore[numerical_cols].head())

# -------------------------
# Task 5: PCA only if correlation exists
# -------------------------
corr = df_clean[numerical_cols].corr()
print("\nCorrelation Matrix:")
print(corr)

if abs(corr.iloc[0, 1]) > 0.5:
    from sklearn.decomposition import PCA
    pca = PCA(n_components=1)
    pca_result = pca.fit_transform(df_clean[numerical_cols])
    print("\nPCA applied successfully.")
    print("Explained variance ratio:", pca.explained_variance_ratio_)
else:
    print("\nPCA not applied because there is no strong correlation between numerical features.")
```
